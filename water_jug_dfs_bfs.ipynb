```python
class waterJug:
   
    def __init__(self,cap_x,cap_y,dest_x,dest_y):
        # Setting up the Capacity
        self.capacityX=cap_x
        self.capacityY=cap_y
       
        # Setting up the Destiny
        self.destX=dest_x
        self.destY=dest_y
       
        # Setting up the initial state
        self.ini_X=0
        self.ini_Y=0
       
        # Additional Props
        self.graph = {}
        self.final = []
       
    
       
    def filling_X(self,x,y):
        if(x==0): # x < self.capacityX
            x = self.capacityX
            y = y
           
            return x,y
       
        
   
    def filling_Y(self,x,y):
        if(y==0): # y < self.capacityY
            y=self.capacityY
            x=x
           
            return x,y
           
        
           
    def empty_X(self,x,y):
        if(x > 0):
            x = 0
           
            return x,y
           
       
           
    def empty_Y(self,x,y):
        if(y > 0):
            y = 0
           
            return x,y
       
        

    def pour_X_Y(self,x,y):
       
        if( 0 < x+y >= self.capacityY and x > 0):
           
            x = x-(self.capacityY - y)
            y = self.capacityY
           
            return x,y
           
       
   
    def pour_Y_X(self,x,y):
       
        if( 0 < x+y >= self.capacityX and y > 0):
           
            y = y-(self.capacityX-x)
            x = self.capacityX
           
           
            return x,y
           
       
           
    def pour_full_X_Y(self,x,y):
       
        if( 0 < x+y <= self.capacityX and y >= 0):
           
            x = x+y
            y = 0
           
            return x,y
        
           
    def pour_full_Y_X(self,x,y):
       
        if( 0 < x+y <= self.capacityY and x >= 0):
           
            y = x+y
            x = 0
           
            return x,y
           
        
           
           
           
    def checker(self):
        print(self.ini_X,self.ini_Y)
       
   
    # Graph related work
    def addNetwork(self,u1,v1,u2,v2):
        if(self.graph.get((u1,v1)) == None):
            self.graph[(u1,v1)] = []
            self.graph[(u1,v1)].append((u2,v2))
       
        else:
            self.graph[(u1,v1)].append((u2,v2))
           
    def check_graph(self):
        return self.graph
   
    # DFS Technique
    def DFSUtil(self, u,v, visited,l):

        visited.add((u,v))
        l.append((u,v))
        print(u,v," ", end=' ')
     
        if((u,v)==(self.destX,self.destY)):
           
            self.final = [i for i in l]
           
       
       
        if(self.graph.get((u,v))!=None):
            for neighbour in self.graph[(u,v)]:

                if neighbour[0] not in visited and neighbour[1] not in visited:
                    self.DFSUtil(neighbour[0],neighbour[1], visited,l)
                   
        else:
            l.pop()
           
       

    def DFS(self, u,v):
 

        visited = set()
        l=[]

        self.DFSUtil(u,v, visited,l)
           
    def ans_(self):
        return self.final
```

```python
from collections import deque
 
def BFS(a, b, target):
     
    
    m = {}
    isSolvable = False
    path = []
     
    
    q = deque()
     
    
    q.append((0, 0))
 
    while (len(q) > 0):
         
        # Current state
        u = q.popleft()
 
       
        if ((u[0], u[1]) in m):
            continue
 
        if ((u[0] > a or u[1] > b or
             u[0] < 0 or u[1] < 0)):
            continue
 
       
        path.append([u[0], u[1]])
 
        
        m[(u[0], u[1])] = 1
 
        
        if (u[0] == target or u[1] == target):
            isSolvable = True
             
            if (u[0] == target):
                if (u[1] != 0):
                     
                  
                    path.append([u[0], 0])
            else:
                if (u[0] != 0):
 
                    
                    path.append([0, u[1]])
 
            
            sz = len(path)
            for i in range(sz):
                print("(", path[i][0], ",",
                           path[i][1], ")")
            break
 
       
        q.append([u[0], b]) # Fill Jug2
        q.append([a, u[1]]) # Fill Jug1
 
        for ap in range(max(a, b) + 1):
 
            c = u[0] + ap
            d = u[1] - ap
 
            
            if (c == a or (d == 0 and d >= 0)):
                q.append([c, d])
 
            
            c = u[0] - ap
            d = u[1] + ap
 
           
            if ((c == 0 and c >= 0) or d == b):
                q.append([c, d])
         
        
        q.append([a, 0])
         
        
        q.append([0, b])
 
    
    if (not isSolvable):
        print ("No solution")
if __name__ == '__main__':
     
    Jug1, Jug2, target = 4, 3, 2
    print("Path from initial state "
          "to solution state ::")
     
    BFS(Jug1, Jug2, target)
 
```
